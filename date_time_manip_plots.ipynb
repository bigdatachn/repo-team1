```python
# Label encode the categorical column
enum_dict = {}
for i, cat in enumerate(df11.cat_feature.unique()):
    enum_dict[cat] = i
df11["cat_feature_enum"] = df11["cat_feature"].replace(enum_dict)
```

```python
enum_dict
```

```python
df11
```

```python
import numpy as np
fig, ax = plt.subplots(figsize=(8, 4))
sns.scatterplot(data = df11,
                x = "Date", 
                y = "cat_feature_enum", 
                hue = "cat_feature",
                marker = '.',
                linewidth = 0,
                )
ax.set_yticks(np.arange(0, (len(df11.cat_feature.unique()) + 1), 1))
ax.set_yticklabels(df11.cat_feature.unique())
ax.get_legend().remove() # remove legend - it's not necessary here
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.vlines(df11["Date"], 0, df11["cat_feature_enum"])
plt.show()
```

```python
ax.set_xlim([df11.Date.min(), df11.Date.max()])
```

```python
ax.set_xlim(np.array(["2020-04-01", "2020-04-30"],
                      dtype="datetime64"))
```

```python
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %d"));
ax.xaxis.set_minor_locator(mdates.DayLocator())
```
